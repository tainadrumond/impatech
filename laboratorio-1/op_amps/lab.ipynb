```python
import numpy as np

import matplotlib.pyplot as plt

# Define the range for x
x = np.linspace(-10, 10, 400)

# Define the constant y value
y = -10

# Create the plot
plt.plot(x, [y]*len(x), label='y = -10')

# Add title and labels
plt.title('Bode de ganho')
plt.xlabel('freq(Hz)')
plt.ylabel('ganho (dB)')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Define the range for x
x = np.linspace(-10, 10, 400)

# Define the constant y value
y = 180

# Create the plot
plt.plot(x, [y]*len(x), label='y = -10')

# Add title and labels
plt.title('Bode de fase')
plt.xlabel('freq(Hz)')
plt.ylabel('fase (Â°)')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Definir a constante r e c
r = 2164
c = 67e-9

# Calcular w como 10^x
w = 10**x

# Calcular f(w)
f_w = (1 + (r * w * c)**2)**(-1/2)

# Calcular y = 20 * log10(f(w))
y = 20 * np.log10(f_w)

# Criar o grÃ¡fico
plt.plot(np.log10(w), y, label='f(w) = (1 + (rwc)^2)^(-1/2)')

# Adicionar tÃ­tulo e rÃ³tulos
plt.title('Bode de ganho')
plt.xlabel('freq(Hz)')
plt.ylabel('ganho (dB)')

# Adicionar uma legenda
plt.legend()

# Mostrar o grÃ¡fico
plt.show()
```

```python
# Calcular y como arctg(-rwc) em graus
y = np.degrees(np.arctan(-r * w * c))

# Criar o grÃ¡fico
plt.plot(np.log10(w), y, label='y = arctg(-rwc)')

# Adicionar tÃ­tulo e rÃ³tulos
plt.title('Bode de fase')
plt.xlabel('Freq(Hz)')
plt.ylabel('Fase(Â°)')

# Adicionar uma legenda
plt.legend()

# Mostrar o grÃ¡fico
plt.show()
```
