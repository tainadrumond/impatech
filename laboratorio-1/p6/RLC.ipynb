```python
import pandas as pd
import numpy as np

simulation = pd.read_csv('./CircuitLab_simulation_RLC.csv').to_numpy()
bode_RL = pd.read_csv('./bode_RL_juntos.csv').to_numpy()

simulation
```

```python
bode_RL
```

```python
smaller = float('inf')
smaller_i = -1
buffer = []
j = 0
for i in range(len(simulation)):
    diff = abs(simulation[i][0] - bode_RL[j][0])
    if diff < smaller:
        smaller = diff
        smaller_i = i
        continue
    else:
        buffer.append([bode_RL[j][0], simulation[i-1][2]])
        smaller = float('inf')
        j += 1
    
buffer
```

```python
df = pd.DataFrame(buffer, columns=['Freq(Hz)','Phase'])
df
```

```python
df.to_csv('./output.csv', index=False)
```
