```python
import pandas as pd
import numpy as np

simulation = pd.read_csv('./CircuitLab_simulation_RLC.csv').to_numpy()
bode_RLC = pd.read_csv('./bode_RLC_juntos.csv').to_numpy()

simulation
```

```python
bode_RLC
```

```python
smaller = float('inf')
buffer = []
j = 0
for i in range(len(simulation)):
    if j >= len(bode_RLC):
        break
    diff = abs(simulation[i][0] - bode_RLC[j][0])
    if diff < smaller:
        smaller = diff
        continue
    else:
        buffer.append([bode_RLC[j][0], simulation[i-1][2]])
        smaller = float('inf')
        j += 1
    
buffer
```

```python
df = pd.DataFrame(buffer, columns=['Freq(Hz)','Phase'])
df
```

```python
df.to_csv('./output.csv', index=False)
```
