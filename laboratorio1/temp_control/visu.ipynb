```python
import pandas as pd
import altair as alt
```

```python
volt = pd.read_csv('./volt.csv')
volt = volt.reset_index().rename(columns={'index': 'tempo'})
volt
```

```python
chart = alt.Chart(volt).mark_point().encode(
    alt.X('tempo', scale=alt.Scale(zero=False)),
    alt.Y('tensao', scale=alt.Scale(zero=False))
)
chart
```

```python
temp = pd.read_csv('./temp.csv')
temp = temp.reset_index().rename(columns={'index': 'tempo'})
temp
```

```python
chart = alt.Chart(temp).mark_point().encode(
    alt.X('tempo', scale=alt.Scale(zero=False)),
    alt.Y('temp', scale=alt.Scale(zero=False))
)
chart
```
