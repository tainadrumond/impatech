### Item (e)

Precisamos encontrar os intervalos associados com os dois coeficientes da função objetivo.

O ponto ótimo encontrado é (0,2). A restrição ativa nesse vértice é 12x + 3y ≤ 6, cuja inclinação é -4. A função objetivo z = c<sub>1</sub> x + c<sub>2</sub> y tem inclinação -c<sub>1</sub>/c<sub>2</sub>. 

Para que o ponto (0,2) continue ótimo, é necessário que -c<sub>1</sub>/c<sub>2</sub> ≤ -4, o que implica c<sub>1</sub> ≤ 4 c<sub>2</sub>, considerando c<sub>2</sub> positivo. Além disso, c<sub>1</sub> deve ser maior ou igual a zero para que o ponto não deixe de ser ótimo no eixo x. 

Portanto, o intervalo de variação do coeficiente de x é 0 ≤ c<sub>1</sub> ≤ 4 c<sub>2</sub>, com c<sub>2</sub> positivo. 

Fixando os coeficientes originais da função objetivo, c<sub>1</sub> = 2 e c<sub>2</sub> = 1, vemos que o intervalo de variação para c<sub>1</sub> é 0 ≤ c<sub>1</sub> ≤ 4. Se fixarmos c<sub>1</sub> = 2, a desigualdade c<sub>1</sub> ≤ 4 c<sub>2</sub> dá 2 ≤ 4 c<sub>2</sub>, ou seja, c<sub>2</sub> ≥ 0,5. Assim, com c<sub>1</sub> fixo, c<sub>2</sub> pode variar de 0,5 até qualquer valor positivo sem alterar o ponto ótimo.

### Item (f)

Precisamos encontrar os intervalos do lado direito para as três restrições.

No ponto (0,2), a primeira restrição `12x + 3y ≤ b1` (b1 = 6) é ativa, enquanto a segunda restrição `-3x + y ≤ b2` (b2 = 7) e a terceira `y ≤ b3` (b3 = 10) não são ativas.

Para a primeira restrição, qualquer diminuição do RHS abaixo de 6 tornaria o ponto (0,2) inviável, então a diminuição permitida é 0. Aumentando o RHS, a restrição se afasta do ponto ótimo e o ponto permanece viável até que outro vértice se torne mais favorável, o que pelo gráfico permite um aumento para até 21, resultando em b1 ∈ [6,21].

Para a segunda restrição, como o ponto ótimo não está sobre a linha, podemos diminuir o RHS até que a linha toque o ponto ótimo sem inviabilizá-lo, resultando em b2 mínimo = 2. Aumentos não alteram o ponto ótimo, então o aumento permitido é infinito, b2 ∈ [2,∞).

Para a terceira restrição, a lógica é similar à segunda: podemos diminuir até b3 = 2, e aumentos não alteram o ponto ótimo, então b3 ∈ [2,∞).

## Exercício 3

#### Problema:

Nesse exercício, queremos resolver o seguinte problema de otimização de uma carteira de títulos:

Um gerente de portfólio de um banco tem $10 milhões para investir nos seguintes títulos:

| Título | Tipo        | Moody’s | Escala do Banco | Anos até vencimento | Rendimento (%) | Rendimento líquido (%) |
|--------|-------------|---------|-----------------|---------------------|----------------|-------------------------|
| A      | Municipal   | Aa      | 2               | 9                   | 4.3            | 4.3                     |
| D      | Governo     | Aaa      | 1               | 3                   | 4.4            | 2.2                     |

Restrições
1. O total investido em títulos governamentais e de agências deve ser de pelo menos $4 milhões.  
2. A qualidade média do portfólio não pode exceder 1,4 na escala do banco.  
3. A média dos anos até o vencimento não pode ultrapassar 5 anos.  
4. O total investivo em títulos municipais deve ser menor ou igual a $3 milhões.

Objetivo: Maximizar o retorno líquido após impostos do portfólio.  

Se for possível tomar emprestado até $1 milhão a 5,5% antes de impostos, como a seleção mudaria?  


### Item (a)

Queremos obter a solução ótima para o problema. Faremos isso com PuLP.

#### Programa:

Sejam x<sub>A</sub> e x<sub>D</sub> a quantidade investida nos títulos A e D em milhões, respectivamente, e y a quantidade emprestada em milhões de dólares.

Queremos maximizar 

z = 0.043x<sub>A</sub> + 0.022x<sub>B</sub> - 0.0275y

com as restrições:
- y ≤ 1,
- x<sub>A</sub> + x<sub>D</sub> ≤ 10 + y => x<sub>A</sub> + x<sub>D</sub> - y ≤ 10,
- x<sub>A</sub> ≤ 3,
- x<sub>D</sub> ≥ 4,
- (2x<sub>A</sub> + x<sub>D</sub>)/(x<sub>A</sub> + x<sub>D</sub>) ≤ 1.4 => 0.6x<sub>D</sub> - 0.4x<sub>D</sub> ≤ 0
- (9x<sub>A</sub> + 3x<sub>D</sub>)/(x<sub>A</sub> + x<sub>D</sub>) ≤ 5 => 4x<sub>D</sub> - 2x<sub>D</sub> ≤ 0
- x<sub>A</sub> ≥ 0, 
- x<sub>D</sub> ≥ 0,
- y ≥ 0.

#### Solução:

```python
# Criação do problema
ex3_lp = pl.LpProblem("Bond_Portifolio", pl.LpMaximize)

# Variáveis de decisão (em milhões)
ex3_A = pl.LpVariable("A", lowBound=0)
ex3_D = pl.LpVariable("D", lowBound=0)
ex3_Y = pl.LpVariable("Y", lowBound=0)

# Função objetivo
ex3_lp += 0.043*ex3_A + 0.022*ex3_D - 0.0275*ex3_Y, "After_Tax_Yield"

# Restrições
ex3_lp += ex3_Y <= 1, "Borrow_Limit"
ex3_lp += ex3_A + ex3_D - ex3_Y <= 10, "Budget"
ex3_lp += ex3_A <= 3, "Max_Municipal"
ex3_lp += ex3_D >= 4, "Min_Government"
ex3_lp += 0.6*ex3_A - 0.4*ex3_D <= 0, "Quality_Scale"
ex3_lp += 4*ex3_A - 2*ex3_D <= 0, "Years_Maturity"

# Resolver
status = ex3_lp.solve(pl.PULP_CBC_CMD(msg=0))

# Solução
print("Status:", pl.LpStatus[ex3_lp.status])
for var in ex3_lp.variables():
    print(var, "=", var.varValue)
print("z =", pl.value(ex3_lp.objective))
```

Repare que, como `Y = 0.0` a seleção não muda com a possibilidade de pegar empréstimo na taxa de juros estabelecida.

### Item (b)

Vamos obter o shadow price para o limite municipal:

```python
shadow_max_municipal = ex3_lp.constraints["Max_Municipal"].pi
print("Shadow price on the municipal limit =", shadow_max_municipal)
```

Para isso, vamos calcular qual seria a solução ótima se não houvesse o limite municipal:

O preço sombra é de 0.022. Isso indica que, se aumentássemos o orçamento em 1 milhão, teríamos um retorno de 0.022. Para que o empréstimo seja favorável, a taxa de juros não pode ser maior que 0.022.

Um exemplo alterando a taxa de 0.0275 para 0.021:

```python
# Criação do problema
ex3d_lp = pl.LpProblem("Bond_Portifolio", pl.LpMaximize)

# Variáveis de decisão (em milhões)
ex3d_A = pl.LpVariable("A", lowBound=0)
ex3d_D = pl.LpVariable("D", lowBound=0)
ex3d_Y = pl.LpVariable("Y", lowBound=0)

# Função objetivo
ex3d_lp += 0.043*ex3d_A + 0.022*ex3d_D - 0.021*ex3d_Y, "After_Tax_Yield"

# Restrições
ex3d_lp += ex3d_Y <= 1, "Borrow_Limit"
ex3d_lp += ex3d_A + ex3d_D - ex3d_Y <= 10, "Budget"
ex3d_lp += ex3d_A <= 3, "Max_Municipal"
ex3d_lp += ex3d_D >= 4, "Min_Government"
ex3d_lp += 0.6*ex3d_A - 0.4*ex3d_D <= 0, "Quality_Scale"
ex3d_lp += 4*ex3d_A - 2*ex3d_D <= 0, "Years_Maturity"

# Resolver
status = ex3d_lp.solve(pl.PULP_CBC_CMD(msg=0))

# Solução
print("Status:", pl.LpStatus[ex3d_lp.status])
for var in ex3d_lp.variables():
    print(var, "=", var.varValue)
print("z =", pl.value(ex3d_lp.objective))
```

Agora, o valor emprestado deixou de ser 0, porque, com a nova taxa, fazer o empréstimo se tornou vantajoso.

## Exercício 9

#### Problema:

A Candid Camera Company fabrica três tipos de câmeras: Cub, Quickiematic e VIP, com contribuições de $3, $9 e $25 por unidade, respectivamente. O centro de distribuição exige que sejam produzidos ao menos 250 Cubs, 375 Quickiematics e 150 VIPs a cada semana.  

Cada câmera demanda tempos específicos em horas para três etapas: fabricação, montagem e inspeção/embalagem. O Cub requer 0.1, 0.2 e 0.1 horas em cada etapa; o Quickiematic exige 0.2, 0.35 e 0.2 horas; e o VIP consome 0.7, 0.1 e 0.3 horas, respectivamente.  

Há restrições semanais de capacidade: 250 horas para fabricação, 350 horas para montagem e 150 horas para inspeção/embalagem.  

O objetivo é formular este problema como um programa linear que maximiza a contribuição total.  


#### Programa:

Maximizar z = 3x<sub>1</sub> + 9x<sub>2</sub> + 25x<sub>3</sub> com as restrições:
- x<sub>1</sub> ≥ 250 (o centro de distribuição requer pelo menos 250 Cubs),
- x<sub>2</sub> ≥ 375 (o centro de distribuição requer pelo menos 375 Quickiematics),
- x<sub>3</sub> ≥ 150 (o centro de distribuição requer pelo menos 150 VIPs),
- 0.1x<sub>1</sub> + 0.2x<sub>2</sub> + 0.7x<sub>3</sub> ≤ 250 (horas de fabricação disponíveis),
- 0.2x<sub>1</sub> + 0.35x<sub>2</sub> + 0.1x<sub>3</sub> ≤ 350 (horas de montagem disponíveis),
- 0.1x<sub>1</sub> + 0.2x<sub>2</sub> + 0.3x<sub>3</sub> ≤ 150 (horas de inspeção, teste e embalagem disponíveis).

#### Solução:

```python
# Criação do problema
ex9_lp = pl.LpProblem("Camera_Production", pl.LpMaximize)

# Variáveis de decisão
ex9_x1 = pl.LpVariable("Cub", lowBound=250, cat="Integer")  # quantidade de câmeras Cub produzidas
ex9_x2 = pl.LpVariable("Quickiematic", lowBound=375, cat="Integer")  # quantidade de câmeras Quickiematic produzidas
ex9_x3 = pl.LpVariable("VIP", lowBound=150, cat="Integer")  # quantidade de câmeras VIP produzidas

# Função objetivo
ex9_lp += 3*ex9_x1 + 9*ex9_x2 + 25*ex9_x3, "Contribution"

# Restrições
ex9_lp += 0.1*ex9_x1 + 0.2*ex9_x2 + 0.7*ex9_x3 <= 250, "Manufacture"
ex9_lp += 0.2*ex9_x1 + 0.35*ex9_x2 + 0.1*ex9_x3 <= 350, "Assemble"
ex9_lp += 0.1*ex9_x1 + 0.2*ex9_x2 + 0.3*ex9_x3 <= 150, "Inspect"

# Resolver
status = ex9_lp.solve(pl.PULP_CBC_CMD(msg=0))
# Solucão
print("Status:", pl.LpStatus[ex9_lp.status])
for var in ex9_lp.variables():
    print(var, " = ", var.varValue)
print("z = ", pl.value(ex9_lp.objective))
```

- 3x<sub>1</sub> + 4x<sub>2</sub> + 0x<sub>3</sub> + 2x<sub>4</sub> + 4x<sub>5</sub> ≤ 1600/12 (acessórios).

#### Solução:

```python
# Criação do problema
ex10_lp = pl.LpProblem("Leather_Goods_Factory", pl.LpMaximize)

# Variáveis de decisão
x1 = pl.LpVariable("Bag1", lowBound=0, upBound=100, cat="Integer")
x2 = pl.LpVariable("Bag2", lowBound=0, upBound=50,  cat="Integer")
x3 = pl.LpVariable("Bag3", lowBound=0, upBound=90,  cat="Integer")
x4 = pl.LpVariable("Bag4", lowBound=0, upBound=70,  cat="Integer")
x5 = pl.LpVariable("Bag5", lowBound=0, upBound=30,  cat="Integer")

# Função objetivo
ex10_lp += 30*x1 + 40*x2 + 45*x3 + 25*x4 + 60*x5, "Contribution"

# Restrições de mão de obra
ex10_lp += 3*x1 + 4*x2 + 2*x3 + 2*x4 + 5*x5 <= 700, "Click"
ex10_lp += 8*x1 + 3*x2 + 2*x3 + 1*x4 + 4*x5 <= 600, "Pare"
ex10_lp += 2*x1 + 1*x2 + 3*x4 + 4*x5 <= 400, "Stitch"
ex10_lp += 6*x1 + 2*x3 + 4*x4 + 3*x5 <= 900, "Finish"

# Restrições de materiais
ex10_lp += 4*x2 + 5*x3 + 6*x4 + 2*x5 <= 300/12, "Leather"
ex10_lp += 1*x1 + 7*x3 + 4*x4 + 3*x5 <= 400/12, "Fabric"
ex10_lp += 4*x1 + 7*x2 + 6*x3 + 1*x4 + 3*x5 <= 1000/12, "Backing"
ex10_lp += 2*x1 + 4*x2 + 2*x3 + 1*x4 <= 900/12, "Lining"
ex10_lp += 3*x1 + 4*x2 + 2*x4 + 4*x5 <= 1600/12, "Acessories"

# Resolver
status = ex10_lp.solve(pl.PULP_CBC_CMD(msg=0))

# Solução
print("Status:", pl.LpStatus[ex10_lp.status])
for var in ex10_lp.variables():
    print(var, "=", var.varValue)
print("z =", pl.value(ex10_lp.objective))
```

## Exercício 11

#### Problema:

Uma corporação que produz gasolina e aditivos especiais de óleo compra três tipos de destilados de petróleo, A, B e C.  
A empresa então combina esses três de acordo com especificações de percentuais máximos ou mínimos dos aditivos A ou C em cada mistura.

| Mistura   | Máx % permitido para aditivo A | Min % permitido para aditivo C | Preço de venda $/galão |
|-----------|--------------------------------|--------------------------------|-------------------------|
| Deluxe    | 60%                            | 20%                            | 7.9                     |
| Standard  | 15%                            | 60%                            | 6.9                     |
| Economy   | —                              | 50%                            | 5.0                     |

Os suprimentos dos três aditivos básicos e seus custos são:

| Destilado | Quantidade máxima disponível por dia (galões) | Custo $/galão |
|-----------|----------------------------------------------|---------------|
| A         | 4000                                         | 0.60          |
| B         | 5000                                         | 0.52          |
| C         | 2500                                         | 0.48          |

Mostre como formular um programa linear para determinar a política de produção que maximiza os lucros.

#### Modelagem:

Para esse programa, criaremos as seguintes variáveis:

Misturas:
- y<sub>D</sub>: galões produzidos para a mistura `Deluxe`,
- y<sub>S</sub>: galões produzidos para a mistura `Standard`,
- x<sub>E</sub>: galões produzidos para a mistura `Economy`.

Destilados:
- x<sub>iA</sub>: galões do destilado `A` utilizados na mistura `i` (ex.: x<sub>DA</sub> indica a quantidade de galões do destilado `A` utilizados na mistura `Deluxe`),
- x<sub>iB</sub>: galões do destilado `B` utilizados na mistura `i`,
- x<sub>iC</sub>: galões do destilado `C` utilizados na mistura `i`.

Podemos relacionar esses dois conjuntos de variáveis da seguinte forma:
- y<sub>D</sub> = x<sub>DA</sub> + x<sub>DB</sub> + x<sub>DC</sub>
- y<sub>S</sub> = x<sub>SA</sub> + x<sub>SB</sub> + x<sub>SC</sub>
- y<sub>E</sub> = x<sub>EA</sub> + x<sub>EB</sub> + x<sub>EC</sub>

O retorno líquido das vendas dos galões de cada mistura é:

P = 7.9y<sub>D</sub> + 6.9y<sub>S</sub> + 5.0y<sub>E</sub> = 7.9(x<sub>DA</sub> + x<sub>DB</sub> + x<sub>DC</sub>) + 6.9(x<sub>SA</sub> + x<sub>SB</sub> + x<sub>SC</sub>) + 5.0(x<sub>EA</sub> + x<sub>EB</sub> + x<sub>EC</sub>)

Os custos de produção podem ser expressos por:

C = 0.6(x<sub>DA</sub> + x<sub>SA</sub> + x<sub>EA</sub>) + 0.52(x<sub>DB</sub> + x<sub>SB</sub> + x<sub>EB</sub>) + 0.48(x<sub>DC</sub> + x<sub>SC</sub> + x<sub>EC</sub>)

Portanto, queremos maximizar a função objetivo
z = P - C


Das especificações de porcentagens máximas e mínimas dos destilados na mistura, temos as seguintes restrições:

- x<sub>DA</sub> / y<sub>D</sub> ≤ 0.6
- x<sub>DC</sub> / y<sub>D</sub> ≥ 0.2
- x<sub>SA</sub> / y<sub>S</sub> ≤ 0.15
- x<sub>DC</sub> / y<sub>S</sub> ≥ 0.6
- x<sub>EC</sub> / y<sub>E</sub> ≥ 0.5

Linearizando essas expressões:
- x<sub>DA</sub> - 0.6y<sub>D</sub> ≤ 0
- x<sub>DC</sub> - 0.2y<sub>D</sub> ≥ 0
- x<sub>SA</sub> - 0.15y<sub>S</sub> ≤ 0
- x<sub>DC</sub> - 0.6y<sub>S</sub> ≥ 0
- x<sub>EC</sub> - 0.5y<sub>E</sub> ≥ 0

Aqui, podemos fazer uma transformação nas variáveis `y` para deixar as restrições em função apenas dos `x`. Assim, chegamos nas expressões finais:
- x<sub>DA</sub> - 0.6x<sub>DA</sub> - 0.6x<sub>DB</sub> - 0.6x<sub>DC</sub> ≤ 0
- x<sub>DC</sub> - 0.2x<sub>DA</sub> - 0.2x<sub>DB</sub> - 0.2<sub>DC</sub> ≥ 0
- x<sub>SA</sub> - 0.15x<sub>SA</sub> - 0.15x<sub>SB</sub> - 0.15x<sub>SC</sub> ≤ 0
- x<sub>DC</sub> - 0.6x<sub>SA</sub> - 0.6x<sub>SB</sub> - 0.6x<sub>SC</sub> ≥ 0
- x<sub>EC</sub> - 0.5x<sub>EA</sub> - 0.5x<sub>EB</sub> - 0.5x<sub>EC</sub> ≥ 0

Das restrições dos galões de destilado disponíveis por dia, temos:
- x<sub>DA</sub> + x<sub>SA</sub> + x<sub>EA</sub> ≤ 4000
- x<sub>DB</sub> + x<sub>SB</sub> + x<sub>EB</sub> ≤ 5000
- x<sub>DC</sub> + x<sub>SC</sub> + x<sub>EC</sub> ≤ 2500

E, finalmente, as restrições óbvias: x<sub>ij</sub> ≥ 0 para todo i em {D, S, E} e para todo j em {A, B, C}.

#### Programa:

Maximizar
z = [7.9(x<sub>DA</sub> + x<sub>DB</sub> + x<sub>DC</sub>) + 6.9(x<sub>SA</sub> + x<sub>SB</sub> + x<sub>SC</sub>) + 5.0(x<sub>EA</sub> + x<sub>EB</sub> + x<sub>EC</sub>)] - [0.6(x<sub>DA</sub> + x<sub>SA</sub> + x<sub>EA</sub>) + 0.52(x<sub>DB</sub> + x<sub>SB</sub> + x<sub>EB</sub>) + 0.48(x<sub>DC</sub> + x<sub>SC</sub> + x<sub>EC</sub>)]

com as restrições:
- x<sub>DA</sub> - 0.6x<sub>DA</sub> - 0.6x<sub>DB</sub> - 0.6x<sub>DC</sub> ≤ 0
- x<sub>DC</sub> - 0.2x<sub>DA</sub> - 0.2x<sub>DB</sub> - 0.2<sub>DC</sub> ≥ 0
- x<sub>SA</sub> - 0.15x<sub>SA</sub> - 0.15x<sub>SB</sub> - 0.15x<sub>SC</sub> ≤ 0
- x<sub>DC</sub> - 0.6x<sub>SA</sub> - 0.6x<sub>SB</sub> - 0.6x<sub>SC</sub> ≥ 0
- x<sub>EC</sub> - 0.5x<sub>EA</sub> - 0.5x<sub>EB</sub> - 0.5x<sub>EC</sub> ≥ 0
- x<sub>DA</sub> + x<sub>SA</sub> + x<sub>EA</sub> ≤ 4000
- x<sub>DB</sub> + x<sub>SB</sub> + x<sub>EB</sub> ≤ 5000
- x<sub>DC</sub> + x<sub>SC</sub> + x<sub>EC</sub> ≤ 2500
- x<sub>ij</sub> ≥ 0 para todo i em {D, S, E} e para todo j em {A, B, C}.

#### Solução:

```python
# Criação do problema
ex11_lp = pl.LpProblem("Oil_Corporation", pl.LpMaximize)

# Variáveis de decisão
ex11_DA = pl.LpVariable("MixtureDeluxe_DistillateA", lowBound=0)
ex11_DB = pl.LpVariable("MixtureDeluxe_DistillateB", lowBound=0)
ex11_DC = pl.LpVariable("MixtureDeluxe_DistillateC", lowBound=0)

ex11_SA = pl.LpVariable("MixtureStandard_DistillateA", lowBound=0)
ex11_SB = pl.LpVariable("MixtureStandard_DistillateB", lowBound=0)
ex11_SC = pl.LpVariable("MixtureStandard_DistillateC", lowBound=0)

ex11_EA = pl.LpVariable("MixtureEconomy_DistillateA", lowBound=0)
ex11_EB = pl.LpVariable("MixtureEconomy_DistillateB", lowBound=0)
ex11_EC = pl.LpVariable("MixtureEconomy_DistillateC", lowBound=0)

# Função objetivo
ex11_lp += (
    7.9*(ex11_DA + ex11_DB + ex11_DC)
  + 6.9*(ex11_SA + ex11_SB + ex11_SC)
  + 5.0*(ex11_EA + ex11_EB + ex11_EC)
  - 0.6*(ex11_DA + ex11_SA + ex11_EA)
  - 0.52*(ex11_DB + ex11_SB + ex11_EB)
  - 0.48*(ex11_DC + ex11_SC + ex11_EC)
), "Profit"

# Restrições
ex11_lp += ex11_DA - 0.6*ex11_DA - 0.6*ex11_DB - 0.6*ex11_DC <= 0, "Deluxe_MaxA"
ex11_lp += ex11_DC - 0.2*ex11_DA - 0.2*ex11_DB - 0.2*ex11_DC >= 0, "Deluxe_MinC"
ex11_lp += ex11_SA - 0.15*ex11_SA - 0.15*ex11_SB - 0.15*ex11_SC <= 0, "Standard_MaxA"
ex11_lp += ex11_DC - 0.6*ex11_SA - 0.6*ex11_SB - 0.6*ex11_SC >= 0, "Standard_MinC"
ex11_lp += ex11_EC - 0.5*ex11_EA - 0.5*ex11_EB - 0.5*ex11_EC >= 0, "Economy_MinC"

ex11_lp += ex11_DA + ex11_SA + ex11_EA <= 4000, "Avalability_A"
ex11_lp += ex11_DB + ex11_SB + ex11_EB <= 5000, "Avalability_B"
ex11_lp += ex11_DC + ex11_SC + ex11_EC <= 2500, "Avalability_C"

# Resolver
status = ex11_lp.solve(pl.PULP_CBC_CMD(msg=0))

# Solução
print("Status:", pl.LpStatus[ex11_lp.status])
for var in ex11_lp.variables():
    print(var, "=", var.varValue)
print("z =", pl.value(ex11_lp.objective))
```

## Exercício 12

#### Problema:

A Universal Aviation está atualmente investigando a possibilidade de expandir seu serviço de transporte de passageiros para o negócio de frete aéreo de pequeno porte.  
Com $4.000.000 disponíveis para investir na compra de novas aeronaves de carga bimotoras, a Universal está considerando três tipos de aviões.

- Aeronave **A**: custa $80.000, tem capacidade de carga de 10 toneladas e velocidade de cruzeiro de 350 nós.  
- Aeronave **B**: custa $130.000, transporta 20 toneladas de carga a uma velocidade média de 300 nós.  
- Aeronave **C**: uma versão modificada do modelo B, com provisão para copiloto, velocidade de 300 nós, capacidade de 18 toneladas, e custo de $150.000.  

A aeronave A requer apenas um piloto e, se voar por três turnos, pode operar em média 18 horas por dia.  
Já as aeronaves B e C requerem dois pilotos cada. O modelo B também opera em média de 18 horas por dia em três turnos. O modelo C pode aproveitar 21 horas por dia de uso em turnos de três turnos devido à sua capacidade de carregamento superior.  

O departamento de operações da Universal estima que 150 turnos de piloto por dia estarão disponíveis para as operações.  
As limitações nas instalações de manutenção restringem a compra a no máximo 30 aviões no total.  

As contribuições das aeronaves A, B e C por milha-tonelada são, respectivamente, $1, $8 e $120.  
Além disso, 3.500.000 milhas-toneladas de transporte devem ser completadas a cada dia, mas as entregas não concluídas pela frota própria podem ser subcontratadas para transportadoras externas com a contribuição de $0,20 por milha-tonelada.  

Qual “mix” de aeronaves deve ser adquirido se a empresa deseja maximizar sua contribuição por dia?  
*(Nota: considere que 1 nó = 1 milha por hora).*

#### Modelagem:

- x<sub>A</sub>: Número de aeronave do tipo A
- x<sub>B</sub>: Número de aeronave do tipo B
- x<sub>C</sub>: Número de aeronave do tipo C
- x<sub>O</sub>: Milha-tonelada contratada externamente.

As milhas-toneladas diárias de cada modelo de aeronave são dadas por:

- A: 10 · 350 · 18 = 63000
- B: 20 · 300 · 18 = 108000
- C: 18 · 300 · 21 = 113400

Queremos maximizar a função objetivo

z = 1 · 63000x<sub>A</sub> + 8 · 108000x<sub>B</sub> + 120 · 113400x<sub>C</sub> + 0.20x<sub>O</sub>

Do orçamento disponível para a compra de aeronaves, temos a restrição a seguir:

80000x<sub>A</sub> + 130000x<sub>B</sub> + 150000x<sub>C</sub> ≤ 4000000

O limite de turnos de pilotos disponíveis implica em:

3(x<sub>A</sub> + 2x<sub>B</sub> + 2x<sub>C</sub>) ≤ 150

As restrições das instalações de manutenção:

x<sub>A</sub> + x<sub>B</sub> + x<sub>C</sub> ≤ 30

Demanda de transporte em milhas-toneladas:

63000x<sub>A</sub> + 108000x<sub>B</sub> + 113400x<sub>C</sub> + 0.20x<sub>O</sub> = 3500000

Por fim,
- x<sub>A</sub> ≥ 0 (inteiro)
- x<sub>B</sub> ≥ 0 (inteiro)
- x<sub>C</sub> ≥ 0 (inteiro)
- x<sub>O</sub> ≥ 0

#### Programa:

Da modelagem acima, obtemos o seguinte programa:

Maximizar
z = 1 · 63000x<sub>A</sub> + 8 · 108000x<sub>B</sub> + 120 · 113400x<sub>C</sub> + 0.20x<sub>O</sub>

com as restrições:
- 80000x<sub>A</sub> + 130000x<sub>B</sub> + 150000x<sub>C</sub> ≤ 4000000
- 3(x<sub>A</sub> + 2x<sub>B</sub> + 2x<sub>C</sub>) ≤ 150
- x<sub>A</sub> + x<sub>B</sub> + x<sub>C</sub> ≤ 30
- 63000x<sub>A</sub> + 108000x<sub>B</sub> + 113400x<sub>C</sub> + 0.20x<sub>O</sub> = 3500000
- x<sub>A</sub> ≥ 0 (inteiro)
- x<sub>B</sub> ≥ 0 (inteiro)
- x<sub>C</sub> ≥ 0 (inteiro)
- x<sub>O</sub> ≥ 0

#### Solução:

```python
import pulp as pl

# Criação do problema
ex12_lp = pl.LpProblem("Universal_Aviation", pl.LpMaximize)

# Variáveis de decisão
ex12_A = pl.LpVariable("AicraftA", lowBound=0, cat="Integer")
ex12_B = pl.LpVariable("AircraftB", lowBound=0, cat="Integer")
ex12_C = pl.LpVariable("AircraftC", lowBound=0, cat="Integer")
ex12_O = pl.LpVariable("Outside_Ton_Mile", lowBound=0)

# Função objetivo
ex12_lp += (
    63000*ex12_A
  + 8*108000*ex12_B
  + 120*113400*ex12_C
  + 0.20*ex12_O
), "Contribution"

# Restrições
ex12_lp += 80000*ex12_A + 130000*ex12_B + 150000*ex12_C <= 4000000, "Budget"
ex12_lp += 3*(ex12_A + 2*ex12_B + 2*ex12_C) <= 150, "Pilot_Shifts"
ex12_lp += ex12_A + ex12_B + ex12_C <= 30, "Max_Aircrafts"
ex12_lp += 63000*ex12_A + 108000*ex12_B + 113400*ex12_C + ex12_O == 3500000, "Demand"

# Resolver
status = ex12_lp.solve(pl.PULP_CBC_CMD(msg=0))

# Solução
print("Status:", pl.LpStatus[ex12_lp.status])
for var in ex12_lp.variables():
    print(var, "=", var.varValue)
print("z =", pl.value(ex12_lp.objective))
```
